## Quick graph of Google Trends for school shooting topic 

### pytrend info here: https://github.com/GeneralMills/pytrends

### idea from Khuyen Tran here: https://mathdatasimplified.com/

```python
from pytrends.request import TrendReq
import requests
import plotly.express as px

pytrends = TrendReq(hl='en-US', tz=360)
```

```python
kw_list = ['school shooting', 'school shootings'] #up to 5 keywords can be used in the keyword list
pytrends.build_payload(kw_list, cat=0, timeframe='all', geo='', gprop='')
```

```python
#exploring possible related searches - most seem to be location-dependent
related_queries = pytrends.related_queries()
related_queries
```

```python
#a possible further area to explore - Google Trends as broken down by location
pytrends.interest_by_region(resolution='DMA', inc_low_vol=True, inc_geo_code=False)
```

```python
#plotting the Google Trends for the keywords over time. "all" was used as the time filter
df = pytrends.interest_over_time()

fig = px.line(df, x = df.index, y = ['school shootings', 'school shooting'], 
              title = 'Google Trend data on searches for "school shooting" and "school shootings"',
            labels={"variable": "Search Terms"
                 })
fig.show()
```

```python
fig.write_image("pytrends_schoolshooting.png")
```
